```python
from sklearn.model_selection import cross_val_score

# Define the models
models = {
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "Naive Bayes": GaussianNB(),
    "Decision Tree": DecisionTreeClassifier(random_state=40, max_depth=6)
}

# Loop through the models
for model_name, model in models.items():
    print(f"Model: {model_name}")

    # Cross-validation scores for both training and testing sets
    cv_train_scores = cross_val_score(model, x_train, y_train, cv=10, scoring="accuracy")
    cv_test_scores = cross_val_score(model, x_test, y_test, cv=10, scoring="accuracy")

    # Print train and test accuracies for each fold
    for fold in range(10):
        print(f"Fold {fold+1}: Train Accuracy = {cv_train_scores[fold]}, Test Accuracy = {cv_test_scores[fold]}")

    # Calculate and print average accuracy
    avg_train_accuracy = cv_train_scores.mean()
    avg_test_accuracy = cv_test_scores.mean()
    print(f"Avg Train Accuracy = {avg_train_accuracy}, Avg Test Accuracy = {avg_test_accuracy}")

    print("-" * 50)
```
